```python
import matplotlib.pyplot as plt
import numpy as np


print(plt.style.available)
```

```python
#default style to "seaborn- whitegrid"
plt.style.use('seaborn-v0_8-whitegrid')
#other style will be 'ggplot'
```

```python
#Basic line plot
x = np.linspace(-2,4,100)
y = x**2 - 2 *x +1
plt.plot(x,y)
plt.show()
```

```python
#multiple lines
x = np.linspace(-2,4,100)
y1 = x**3
y2 = np.exp(x)
y3 = np.log(x)
plt.plot(x,y1, label = 'f(x)', color = 'red')
plt.plot(x,y2, label = 'g(x)', color = 'blue')
plt.plot(x,y3,label = 'h(x)', color = 'green')
plt.legend()
plt.show()
```

```python
#Styling lines

x= np.linspace(-10,10,100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5
plt.plot(x,y1, label = 'f(x)', color = 'red', linestyle = '-' )
plt.plot(x,y2, label = 'g(x)', color = '#0000ff', linestyle = '--')
plt.plot(x,y3,label = 'h(x)', color = 'g',linestyle = '-.')
plt.plot(x,y4,label = 'i(x)', color = (0,0,0), linestyle = ':')
plt.legend()
plt.show()
```

```python
#Axes, Limits, and Scaling

x = np.linspace(-10,10,100)
y = 1/ (x**2 + 1)

plt.plot(x,y)

#plt.axis('tight')
plt.axis('equal')
plt.show()

#reversed
x = 1/ (y**2 + 1)
y = np.linspace(-10,10,100)

plt.plot(x,y)

plt.axis('tight')
#plt.axis('equal')
plt.show()
plt
```

```python
#6 Labels and Legends
x = np.linspace(-2,3,100)
y1 = np.sqrt(x)
y2 = x**3

plt.plot(x,y1, label = 'f(x)')
plt.plot(x,y2, label = 'g(x)')
plt.title('f(x) and g(x)')
plt.xlim([0,2])
plt.ylim([0,10])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
#7 Basic Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y,marker = 'o')
plt.show()
plt.scatter(x,y,marker = 'o')
plt.show()
```

```python
#8 Customizing Scatter plots
x = np.linspace(0, 4*np.pi,50)
y =x* np.sin(x)

plt.plot(x,y,markersize = 10, marker= 'o',linewidth = 2, markerfacecolor = 'blue',markeredgecolor = 'green', markeredgewidth = 2 )
```

```python
#9 combined plot
x = np.linspace(-2*np.pi, 2*np.pi,100)
y1 = x* np.cos(x)
x1= np.linspace(-2*np.pi, 2*np.pi,8)
y2 = x1* np.cos(x1)
plt.plot(x,y1)
plt.scatter(x1,y2,marker = '^',color = 'red')
plt.show()
```

```python
#10 custom plot with alrenative style
plt.style.use('ggplot')
x = np.linspace(-10 , 10 , 100)
y1 = np.tan(x)
y2 = np.arctan(x)
plt.plot(x,y2, label = 'Arctan(x)',color = '#FA8072')
plt.ylim(-5,5)
plt.plot(x,y1, label = 'Tan(x)', color = '#87ceeb')
plt.legend()
plt.title('Tan(x) and Arctan(x)')
plt.show()
```

Comments: I chose tan(x) and arctan(x) just because they are inverses of each other. Personally I prefer ggplot just cause I have used R studios and that is the normal package that I always used so I will gravitate to it.

```python
#Styling lines
#first plot using ggplot
plt.style.use('ggplot')
x= np.linspace(-5,5,100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5
plt.ylim(-50,50)
plt.plot(x,y1, label = 'f(x)', color = 'red', linestyle = '-' )
plt.plot(x,y2, label = 'g(x)', color = '#0000ff', linestyle = '--')
plt.plot(x,y3,label = 'h(x)', color = 'g',linestyle = '-.')
plt.plot(x,y4,label = 'i(x)', color = (0,0,0), linestyle = ':')
plt.legend()
plt.show()
#second plot using Seaborn
plt.style.use('seaborn-v0_8-whitegrid')
x= np.linspace(-5,5,100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5
plt.ylim(-50,50)
plt.plot(x,y1, label = 'f(x)', color = 'red', linestyle = '-' )
plt.plot(x,y2, label = 'g(x)', color = '#0000ff', linestyle = '--')
plt.plot(x,y3,label = 'h(x)', color = 'g',linestyle = '-.')
plt.plot(x,y4,label = 'i(x)', color = (0,0,0), linestyle = ':')
plt.legend()
plt.show()\
#third plot using default style
plt.style.use('default')
x= np.linspace(-5,5,100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5
plt.ylim(-50,50)
plt.plot(x,y1, label = 'f(x)', color = 'red', linestyle = '-' )
plt.plot(x,y2, label = 'g(x)', color = '#0000ff', linestyle = '--')
plt.plot(x,y3,label = 'h(x)', color = 'g',linestyle = '-.')
plt.plot(x,y4,label = 'i(x)', color = (0,0,0), linestyle = ':')
plt.legend()
plt.show()
```

The first thing to note would be the default style does not have any grind lines. I think this would be the last option when it comes to chosing a style because of the lack of grid lines. Then when comparing the seaborn and ggplot I think ggplot would be better just cause the gray background makes it more visaully appealing.
