```python
#part 3, Task 4
class1= titanic[titanic['class'] == 'First']['fare']
class2= titanic[titanic['class'] == 'Second']['fare']
class3= titanic[titanic['class'] == 'Third']['fare']

#histogram to compare fare distributions across different passenger classes
plt.hist([class1, class2, class3], bins=15, alpha=0.5, label=['Class 1', 'Class 2', 'Class 3'])
plt.grid(axis = 'y',alpha=0.75)
plt.xlabel('Fare')
plt.legend()
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
#KDE to compare fare distributions across different passenger classes
sns.kdeplot(titanic[titanic['class'] == 'First']['fare'])
sns.kdeplot(titanic[titanic['class'] == 'Second']['fare'])
sns.kdeplot(titanic[titanic['class'] == 'Third']['fare'])
plt.grid(axis = 'y',alpha=0.75)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.legend(['Class 1', 'Class 2', 'Class 3'])
plt.xlim(0,500)
plt.show()
```

```python
#Part 4 Task 5:
#KDE plots to compare the age distributions of survivors and non-survivors.

sns.kdeplot(titanic[titanic['survived'] == 1]['age'])
sns.kdeplot(titanic[titanic['survived'] == 0]['age'])
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(['Survived', 'Not Survived'])
plt.xlim(0,100)
plt.show()
```

# Part 4 Task 5: What trends do you notice in the age distribution of survivors compared to those who did not survive? How might age have impacted survival rates?

For the most part it looks about the same until you look at the younger ages, seems like children and teens were more likely to survive. This could be because Children were prioritized to save.

```python
#Part 4  Task 6
#Create a pyramid-style histogram (also known as a population pyramid) for age, showing survival data split by sex. If possible, attempt to find or create one.

#age_male =titanic[titanic['sex'] == 'male']['age']
#age_female = titanic[titanic['sex'] == 'female']['age']
survived_age_male = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 1)]['age']
survived_age_female = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)]['age']


plt.style.use('ggplot')
plt.hist(survived_age_male, bins=30, alpha=0.5, label='Male', color = 'blue',orientation ='vertical', weights =[-1]*len(survived_age_male))
plt.hist(survived_age_female, bins=30, alpha=0.5, label='Female', color = 'red',orientation ='vertical')
plt.xlabel('Age')
plt.yticks([-20,-15,-10,-5,0,5,10,15,20],[20,15,10,5,0,5,10,15,20])
plt.title('Age Distribution by Survival and Sex')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.legend()
plt.show()
```
