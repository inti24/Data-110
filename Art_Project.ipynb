The data set that I chose was about "Smoking-Attributable Mortality, Morbidity, and Economic Costs (SAMMEC) - Smoking-Attributable Mortality". I connected with this data set because, as a former smoker, I wanted to know just how many deaths could be attributed to smoking. So I made a design to mimic a cigarette butt and each smoke line coming from the cigarette would represent the total number of SAM (Smoking-Attributable Mortatlity) by state. I made the lines relative to each other and added small tick marks at the bottom of the piece to indicate that passing each tick mark is 10,000 SAMs. I made the tick marks small, not to get in the way of the art, because I just wanted people to see just how long some of the lines are, to see how many people die due to smoking.
