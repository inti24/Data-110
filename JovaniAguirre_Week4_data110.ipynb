```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a data frame for crimes in 2023
crimesdata= {
             'Offense' :  ["Aggravated Assault","Burglary","Larceny","Motor Vehicle Theft","Murder","Rape (revised definition)", "Robbery"],
              'Offesnes per 100,000 people (2023)' : [264.1, 250.7,1347.2,318.7,5.7,38.0,66.5]  ,
             'Change in Offenses per 100,000 People (2022- 2023)':[-3.3, -8.1, -4.9,12,-12,-9.8,]
}

crimes = pd.DataFrame(crimesdata)
crimes
```

```python
#make a sorted data frame to make bar graph look better

crimes_sorted = crimes.sort_values(by='Offesnes per 100,000 people (2023)', ascending=True)
crimes_sorted
```

```python
#Task 1
plt.style.use('ggplot') #to have a grid line but also I prefer ggplot

x = crimes_sorted['Offense']
y = crimes_sorted['Offesnes per 100,000 people (2023)']
plt.title('Offenses per 100,000 people (2023)')
plt.ylabel('Offenses')
plt.xlabel('Offenses per 100,000 people (2023)')
plt.barh(x,y)
```

```python
#Task 2
x=crimes_sorted['Offense']
y=crimes_sorted['Change in Offenses per 100,000 People (2022- 2023)']

plt.title('Change in Offenses per 100,000 People (2022- 2023)')
plt.ylabel('Offenses')
plt.xlabel('Change in Offenses per 100,000 People (2022- 2023)')
plt.barh(x,y, color=['red','blue','green','black','brown','pink','purple'])
```

# Task 2 Additional Instuctions:
Absolute numbers can be missleading because it can give people the wrong impression. In this case if you only saw the total amount of crimes then someone might come to the conclusion that crime is at an all time high if they knew nothing about the numbers in the beginning. But if they were to see the change of crime rate then they could see that most crimes are decreasing. They would also get more information about certain types of crime, in this case they could see that motor vehicle theft is on the rise.

```python
#Task 2 Additional instuctions
#next iteration would be to sort the bar chart so that it looks better
#sorting the data frame
crimes_sorted2 = crimes.sort_values(by='Change in Offenses per 100,000 People (2022- 2023)', ascending=True)
crimes_sorted2
```

```python
#second iteration
x=crimes_sorted2['Offense']
y=crimes_sorted2['Change in Offenses per 100,000 People (2022- 2023)']

plt.title('Change in Offenses per 100,000 People (2022- 2023)')
plt.ylabel('Offenses')
plt.xlabel('Change in Offenses per 100,000 People (2022- 2023)')
plt.barh(x,y, color=['red','blue','green','black','brown','pink','purple'])
```

```python
#3rd iteration
#remove some of the grid lines
x=crimes_sorted2['Offense']
y=crimes_sorted2['Change in Offenses per 100,000 People (2022- 2023)']
plt.grid(axis = 'y')
plt.title('Change in Offenses per 100,000 People (2022- 2023)')
plt.ylabel('Offenses')
plt.xlabel('Change in Offenses per 100,000 People (2022- 2023)')
plt.barh(x,y, color=['red','blue','green','black','brown','pink','purple'])
```
